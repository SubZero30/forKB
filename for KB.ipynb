```
import pandas as pd
```

```
df = pd.read_csv('sales_dataset.csv')
```

```
df.columns
```

```
df1 = df.loc[:,['Item Type','Sales Channel','Region']]
```

```
df1.shape
```

```
df2 = df1.copy(deep=True)
df2.columns
```

```
#test case 
#rows = Item Type,sales channel
#columns = Region
#values = empty
# here we calculate frequency of occurrence of Region (which is a column)
```

```
df2['frequency'] = [1]*df2.shape[0]
```

```
df3 = df2.groupby(['Item Type', 'Sales Channel', 'Region']).agg({'frequency':'count'})
```

```
df3
```

```
df4 = df3.unstack(level=-1)
```

```
df4
```

```
df4 = df4.stack(level=0)
```

```
df4
```

```
df4 = df4.unstack(level=-1)
```

```
df4
```

```
#another example
#rows = Item type, Region
#columns = sales channel, order priority
#values = empty
# here we calculate frequency of occurrence of ['Sales Channel','Order Priority'] (both are columns)
```

```
df5 = df.loc[:,['Item Type','Region','Sales Channel','Order Priority']]
df5['frequency'] = [1]*df5.shape[0]
df5 = df5.groupby(['Item Type','Region','Sales Channel','Order Priority']).agg({'frequency':'count'})
df5 = df5.unstack(['Sales Channel','Order Priority'])
df5 = df5.stack(level=0)
df5 = df5.unstack(level=-1)
```
